# Cross validation and learning curve

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import validation_curve
from sklearn.model_selection import learning_curve
from sklearn.neighbors import KNeighborsClassifier
```

```python
bc = load_breast_cancer()
```

```python
x, y = bc.data, bc.target
learner = KNeighborsClassifier()
param_range = range(2, 6)
train_scores, test_scores = validation_curve(learner, x, y, 
                                             param_name = 'n_neighbors', 
                                             param_range = param_range, 
                                             cv = 3,
                                             scoring = 'accuracy')

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
plt.figure()
plt.title('Validation curves')
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha = 0.3, 
                 color = 'C1')
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha = 0.1, 
                 color = 'C0')
plt.plot(param_range, train_scores_mean, 'o-', color = 'C1', label = 'Training score')
plt.plot(param_range, test_scores_mean, 'o-', color = 'C0', label = 'Cross -Validation score')
plt.xticks(param_range)
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.show()
```

```python
x, y = bc.data, bc.target
learner = KNeighborsClassifier()
train_sizes = [50,100,150,200,250,300]
train_sizes, train_scores, test_scores = learning_curve(learner, x, y, 
                                                         train_sizes = train_sizes,
                                                        cv = 10)

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
test_scores_mean
```

```python
plt.figure()
plt.title('Learning curves')
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha = 0.3, 
                 color = 'C1')
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, 
                test_scores_mean + test_scores_std, alpha = 0.1, 
                color = 'C0')
plt.plot(train_sizes, train_scores_mean, 'o-', color = 'C1', label = 'Training score')
plt.plot(train_sizes, test_scores_mean, 'o-', color = 'C0', label = 'Cross -Validation score')
plt.xticks(train_sizes)
plt.xlabel('Size of training set')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.show()
```

# Hard voting

```python
from sklearn import datasets, linear_model, svm, neighbors
from sklearn.metrics import accuracy_score
import numpy as np

breast_cancer = datasets.load_breast_cancer()
x, y = breast_cancer.data, breast_cancer.target
```

```python
learner_1 = neighbors.KNeighborsClassifier(n_neighbors = 5)
learner_2 = linear_model.Perceptron(tol = 1e-2, random_state = 2)
learner_3 = svm.SVC(gamma = 0.001)
```

```python
test_sample = 100
x_train, y_train = x[:-test_sample], y[:-test_sample]
x_test, y_test = x[-test_sample:], y[-test_sample:]

learner_1.fit(x_train, y_train)
learner_2.fit(x_train, y_train)
learner_3.fit(x_train, y_train)
```

```python
predictions_1 = learner_1.predict(x_test)
predictions_2 = learner_2.predict(x_test)
predictions_3 = learner_3.predict(x_test)
```

```python
hard_predictions = []
for i in range(test_sample):
    counts = [0 for _ in range(2)]
    counts[predictions_1[i]] = counts[predictions_1[i]] + 1
    counts[predictions_2[i]] = counts[predictions_2[i]] + 1
    counts[predictions_3[i]] = counts[predictions_3[i]] + 1
    final = np.argmax(counts)
    hard_predictions.append(final)
```

```python
print('L1:', accuracy_score(y_test, predictions_1))
print('L1:', accuracy_score(y_test, predictions_2))
print('L1:', accuracy_score(y_test, predictions_3))
print('-'*30)
print('hard voting:', accuracy_score(y_test, hard_predictions))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('seaborn-paper')
```
